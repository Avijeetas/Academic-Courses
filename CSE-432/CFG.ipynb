```python

CFG ={}
while True:
    pl=input()
    if pl =="endp":
        break
    pr=list(input("-->").split('/'))
    CFG[pl]=pr;
print(CFG)

#CFG ={'S': ['ABC'], 'A': ['aA', 'a'], 'B': ['Bb', 'b'], 'C': ['Ac']}

def cfg(word,checkWord):
    if word ==checkWord:
        return True
    l=len(word)
    if l>len(checkWord):
        return False
    for i in range(l):
        if word[i] in CFG:
            
            for j in range(len(CFG[word[i]])):
                wrd = word[:i]+CFG[word[i]][j]+word[i+1:]
                print(wrd)
                if  cfg(wrd,checkWord):
                    return True
while True:
    
    check = input("Input String:")
    if check == "exit":
        break
    if cfg(CFG["S"][0],check): 
        print("-------------------->POSSIBLE<------------------")
    else:
        print("-------------------->NOT POSSIBLE<------------------")
    

        
    
```
